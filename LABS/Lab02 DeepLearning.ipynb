1. Implement the step function as an activation function in Python using both the math library and TensorFlow. Test it with different input values and visualize its output using a simple plot.

```python
import math
import matplotlib.pyplot as plt
def step_function(x):
    return 1 if x > 0 else 0
def math_activation_func(a,b,function,activation_name):
    inputs = [x for x in range(a,b)]
    output = [function(x) for x in inputs]
    plt.plot(inputs,output)
    plt.title(f"{activation_name} Activation using Math")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid()
    plt.show()
math_activation_func(-5,10,step_function,'Step Function')
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
def step_function_tf(x):
    return tf.where(x > 0, tf.ones_like(x),tf.zeros_like(x))
x = tf.placeholder(tf.float32, name='x')
output = step_function_tf(x)
def tensor_session(a,b,output,activation_name):
    with tf.Session() as sess:
        input_values = inputs = [x for x in range(a,b)]
        output = sess.run(output,feed_dict={x:input_values})
        print(f"{activation_name} of ", input_values, " is " , output )
    plt.plot(input_values,output)
    plt.title(f"{activation_name} Activation using TensorFlow")
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.grid()
    plt.show()
tensor_session(-5,10,output,'Step Function')
```

2. Create a Python function for the ReLU (Rectified Linear Unit) activation function using both the math library and TensorFlow. Test it with a range of input values and plot its output.

```python
def relu(x):
    return max(0,x)

math_activation_func(-10,10,relu,'Relu')
```

```python
x = tf.placeholder(tf.float32, name='x')
output_relu = tf.nn.relu(x)
tensor_session(-10,10,output_relu,'Relu')
```

3. Write a Python code to implement the Sigmoid activation function using both the math library and TensorFlow. Visualize its output and discuss its characteristics compared to other activation functions.

```python
def sigmoid(x):
    return 1/(1 + math.exp(-x))
math_activation_func(-6,8,sigmoid,'Sigmoid')
```

```python
x = tf.placeholder(tf.float32, name='x')
sigmoid_output = tf.sigmoid(x)
tensor_session(-6,8,sigmoid_output,'Sigmoid')
```

4. Develop a Python function for the Tanh (Hyperbolic Tangent) activation function using both the math library and TensorFlow. Test it with various input values and plot its output to observe its behavior.

```python
def tanH(x):
    return ((math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x)))
math_activation_func(-4,10,tanH,'TanH')
```

```python
x = tf.placeholder(tf.float32,name='x')
tanh_output = tf.keras.activations.tanh(x)
tensor_session(-4,10,tanh_output,'Tanh')
```

5. Implement the Leaky ReLU activation function in Python using both the math library and TensorFlow. Experiment with different leakage coefficients and plot the function's output.

```python
def leaky_relu(x):
    return max((0.1 * x),x)
math_activation_func(-10,10,leaky_relu,'Leaky Relu')
```

```python
x = tf.placeholder(tf.float32,name='x')
output = tf.nn.leaky_relu(x, alpha=0.1)
tensor_session(-10,10,output,'Leaky Relu')
```

6. Create a Python function for the Softmax activation function using both the math library and TensorFlow

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
def softmax_math(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x, axis=0)
x = np.linspace(-5, 5, 100)
softmax_math_result = softmax_math(x)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, softmax_math_result)
plt.title('Softmax using Math Library')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
softmax_tf_result = tf.nn.softmax(x)
with tf.compat.v1.Session() as sess:
    softmax_tf_result_value = sess.run(softmax_tf_result)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
plt.plot(x, softmax_tf_result_value)
plt.title('Softmax using TensorFlow')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```
